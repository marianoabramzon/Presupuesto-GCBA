# En este proyecto se realizara un analisis del presupuesto 2022 de GCBA

```python
# importacion de librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lectura del dataset
df = pd.read_csv('presugcba.csv')
df.head(2)
```

Las columnas que presenta el archivo son



* anio
* jur_desc
* fin_desc
* fun_desc
* inciso_desc
* ppal_desc 
* ff_desc 
* eco_desc 
* sancion 
* vigente 
* definitivo
* devengado

```python
# agrupo por año
totalaño = df.groupby('anio').sum()
totalaño
```

```python
# grafico la evolucion año a año

plt.ylabel("Presupuesto")
plt.xlabel("Año")
plt.bar(df["anio"],df["sancion"],color="blue")
plt.show()
```

Se puede observar un crecimiento de tipo Exponencial en el presupuesto
